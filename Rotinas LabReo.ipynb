```python
########### FUNÇÕES GRÁFICAS

def initializePlot(xlim,ylim,title,xtitle,ytitle): # Inicializa gráfico com configurações pré-setadas
    fig.update_layout(
        autosize=False,
        width=800,
        height=800,
        template="plotly_white",
        xaxis_showgrid=True, yaxis_showgrid=True,
        font=dict(
        #family="Courier New, monospace",
        size=15
    )
    )
    fig.update_xaxes(showline=True,ticks="inside",tickwidth=2, linewidth=2, linecolor='black', mirror=True,title_text=xtitle)
    fig.update_yaxes(showline=True,ticks="inside",tickwidth=2, linewidth=2, linecolor='black', mirror=True,title_text=ytitle)
    fig.update_yaxes(range=ylim)
    fig.update_xaxes(range=xlim)
    fig.update_layout(title=title, title_x=0.5)
    fig.update_layout(legend=dict(
        yanchor="top",
        y=0.99,
        xanchor="left",
        x=0.01
    ))
    

def addFlowCurve(x,y,xlim,leg,cor,symbol='circle-open',f=2): # Adiciona curva de escoamento ao gráfico inicializado
    X = []
    Y = []
    for i in range(0,len(x)):
        if x[i]>xlim[0] and x[i]<xlim[1]:
            X.append(x[i])
            Y.append(y[i])
    fig.add_trace(go.Scatter(x=X[1:len(x):f],y=Y[1:len(y):f],mode='markers',marker=dict(color=cor,symbol=symbol,size=10),name=leg))

```
